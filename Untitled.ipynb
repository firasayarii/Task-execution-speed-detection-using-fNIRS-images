```python
import pandas as pd
```

```python
sample=pd.read_excel("TestConditions.xlsx")
```

```python
sample.head()
```

```python
# Créer une liste de noms de colonnes pour un seul canal
channel_columns = ['min_channel', 'max_channel', 'mean_channel', 'std_channel', 
                   'median_channel', 'var_channel', 'kurtosis_channel', 
                   'skewness_channel', 'peak2peak_channel']

# Créer une liste de noms de colonnes pour tous les canaux
all_columns = []
for i in range(1, 15):
    all_columns.extend([f"{name}_{i}" for name in channel_columns])
```

```python
features = pd.read_excel("features.xlsx", header=None, names=all_columns)
```

```python
features
```

```python
df=pd.concat([sample,features],axis=1)
```

```python
pd.set_option('display.max_columns', None)
df.head(20)
```

```python
output_file = 'Sample.xlsx'
df.to_excel(output_file)
```
